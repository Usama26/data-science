```python
import pandas as pd
```

```python
dfA = df[0]
```

1. Renaming Columns
2. Remove NaN values
3. Removing paranthesis
4. Reorder the columns
5. In NET volume, remove paranthesis and place - sign
6. Data Validation: Summing up TOTAL and cross check with the last row values

```python
names = list(dfA.columns)
```

```python
names
```

```python
dfA.set_axis(['CLIENT_TYPE', 'MARKET_TYPE', 'BUY_VOLUME', 'BUY_VALUE', 'SELL_VOLUME' ,'SELL_VALUE','NET_VOLUME',
             'NET_VALUE','USD'], axis=1, inplace=True)
```

```python
dfA.head()
```

```python
dfA.fillna("umair updated", inplace = True)
```

```python
dfA.head()
```

```python
def removingParanthesis(dfA):
    dfA[4] = str(dfA[4]).replace("("," ").replace(")"," ").strip()
    return dfA[4]
```

```python
sellVolUpdated = pd.DataFrame(list(map(removingParanthesis, dfA.values.tolist())))
```

```python
sellVolUpdated.head()
```

```python
dfA = dfA.join(sellVolUpdated)
```

```python
dfA
```

```python
dfA.drop("SELL_VOLUME", axis=1,inplace=True)
dfA.rename(columns={0:'SELL_VOLUME_UPDATED'}, inplace=True)
dfA = dfA[['CLIENT_TYPE', 'MARKET_TYPE', 'BUY_VOLUME', 'BUY_VALUE', 'SELL_VOLUME_UPDATED' ,'SELL_VALUE',\
           'NET_VOLUME','NET_VALUE','USD']]
```

```python
dfA
```

```python
def removingParanAndReplaceWithHyphen(dfA):
    dfA[6] = str(dfA[6]).replace("(",'-').replace(")"," ").replace(',','').strip()
    return dfA[6]
```

```python
netVolUpdated = pd.DataFrame(list(map(removingParanAndReplaceWithHyphen, dfA.values.tolist())))
```

```python
netVolUpdated
```

```python
dfA = dfA.join(netVolUpdated)
```

```python
dfA
```

```python
dfA.drop("NET_VOLUME", axis=1,inplace=True)
dfA.rename(columns={0:'NET_VOLUME_UPDATED'}, inplace=True)
dfA = dfA[['CLIENT_TYPE', 'MARKET_TYPE', 'BUY_VOLUME', 'BUY_VALUE', 'SELL_VOLUME_UPDATED' ,'SELL_VALUE',\
           'NET_VOLUME_UPDATED','NET_VALUE','USD']]
```

```python
dfA.head(10)
```

```python
total = 0
for i in range(0,len(dfA)):
    total = total + int(dfA['NET_VOLUME_UPDATED'][i])
    if dfA['CLIENT_TYPE'][i] == "umair updated":
        print("total = ", total)
        dfA['NET_VOLUME_UPDATED'][i] = total
        break
```

```python
dfA.head(10)
```

```python
dfB = dfA.head(10)
```

```python
total = 0
for i in range(0,len(dfB)):
    if dfB['CLIENT_TYPE'][i] == "umair updated":
        total = dfB['BUY_VOLUME'][i] + total
dfB.loc[-1] = ['NaN', 'NaN', total, 'NaN', 'NaN', 'NaN', 'NaN', 'NaN', 'NaN']
```

## Web Scraping Through Selenium & BeautifulSoup

```python
from bs4 import BeautifulSoup 
import requests
```

```python
data = requests.get("https://nation.com.pk/")
```

```python
soup = BeautifulSoup(data.content, 'html.parser')
```

```python
soup
```

```python
type(soup)
```

```python
a = soup.find_all('ul', {'class':'list-latestcat'})
```

```python
for i in range(0,len(a)):
    s = a[i].find('a')
    print("title: ", s.text)
    print("link: ", s.get('href') + '\n')
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
